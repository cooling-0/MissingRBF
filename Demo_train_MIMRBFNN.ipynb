```python
import pandas as pd
import numpy as np
import torch
from utils import Dataset
from models import MultiRBFnnTime
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
humanth = 0
numth = 0
Tagth = 0
input_data, target, lossth, target_ground = Dataset(humanth, numth, Tagth, device)
```

```python
train_masking = torch.where(target.isnan(), 0.0 , 1.0)
real_masking = torch.where(target_ground.isnan(), 0.0 , 1.0)
missing_index = real_masking - train_masking
torch.sum(missing_index)
```

```python
model = MultiRBFnnTime(target.size(0), add_rbf_num = 20, device=device)
```

```python
model.train(input_data = input_data, target = target, epochs = 3000,
                    lr = 1e-3, loss_th=lossth, lr_change_th=lossth)
```

```python
import matplotlib.pyplot as plt
pred = model.pred(input_data)[1]
for i in range(3):
    print(torch.sum(missing_index[i]))
    plt.scatter(input_data[missing_index[i] != 0].cpu().detach().numpy(), target_ground[i][missing_index[i] != 0].cpu().detach().numpy(), c = 'blue')
    plt.scatter(input_data[missing_index[i] != 0].cpu().detach().numpy(), pred[i][missing_index[i] != 0].cpu().detach().numpy(), c = 'red')
    plt.show()
```

```python
print(torch.mean(torch.abs(target_ground[(missing_index != 1)] -  pred[(missing_index != 1)])))
print(torch.mean(torch.abs(target_ground[(train_masking != 1)] -  pred[(train_masking != 1)])))
```
